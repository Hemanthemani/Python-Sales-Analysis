```python
import pandas as pd
```

Download Sales Data from Gethub

```python
Sales_Jan=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_January_2019.csv')
Sales_Feb=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_February_2019.csv')
Sales_Mar=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_March_2019.csv')
Sales_Apr=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_April_2019.csv')
Sales_May=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_May_2019.csv')
Sales_Jun=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_June_2019.csv')
Sales_Jul=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_July_2019.csv')
Sales_Aug=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_August_2019.csv')
Sales_Sep=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_September_2019.csv')
Sales_Oct=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_October_2019.csv')
Sales_Nov=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_November_2019.csv')
Sales_Dec=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_December_2019.csv')
```

Combine all Data Files


```python
all_data=[Sales_Jul,Sales_Feb,Sales_Mar,Sales_Apr,Sales_May,Sales_Jun,Sales_Jul,Sales_Aug,Sales_Sep,Sales_Oct,Sales_Nov,Sales_Dec]
```

```python
df=pd.concat(all_data)
```

```python
df
```

**Clean up the Data**

```python
df.isna().sum()
```

```python
df.dropna(how='all',inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# First we convert Column to string then we extract first two char and then convert to int  
df['month']=df['Order Date'].str[0:2]
```

```python
df=df[df['month']!='Or']
```

```python
df['month']=df['month'].astype('int32')
df.head()
```

Add Sales Column

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('int32')
df['Price Each']=df['Price Each'].astype('float')
```

```python
df['Total Sales']=df['Price Each']*df['Quantity Ordered']
```

```python
df
```

**Add City Column**

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
df.head()
```

### **Data Exploration!**

Q1 What was the best month for Sales? How much was that earned?

```python
result=df.groupby(df['month']).sum()
result
```

```python
import matplotlib.pyplot as plt 

months=range(1,13)
print(months)

plt.bar(months,result['Total Sales'])
plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales in USD $')
plt.show()
```

**Q2: What city sold most product?**

```python
results=df.groupby(['City']).sum()
results
```

```python

cities=[city for city,df in df.groupby('City')]

plt.bar(cities,results['Total Sales'])
plt.xticks(cities,rotation='vertical',size=10)
plt.xlabel('City')
plt.ylabel('Sales in USD $')
plt.show()
```

**Q3: What time should we display advertisements to maximize likehood of customer's buying the Product**

```python
# Add hour column
df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
df['Minute'] = pd.to_datetime(df['Order Date']).dt.minute
df['Count'] = 1
df.head()
```

```python
hours=[hour for hour,df in df.groupby('Hour')]

plt.plot(hours,df.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Time')
plt.ylabel('No of Order')
plt.grid()
plt.show()
```

**Q4: What Product sold most often together ?**

```python
df1=df[df['Order ID'].duplicated(keep=False)]
df1['Grouped']=df1.groupby('Order ID')['Product'].transform(lambda x :','.join(x))
df1=df1[['Order ID','Grouped']].drop_duplicates()
df1.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter 

count=Counter()

for row in df1['Grouped']:
  row_list=row.split(',')
  count.update(Counter(combinations(row_list,1)))


for key,value in count.most_common(10):
  print(key,value)
```
